```python
#dropping first column
df =data.drop(columns=['Unnamed: 0'])
```

Normalising the dataset

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
 # Normalisation
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

Applying PCA technique

```python
# Perform PCA , reducing the dimensionality
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

# Create a DataFrame for PCA results
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])

```

```python
# Visualize the PCA results
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', data=df_pca)
plt.show()
```

## K Means Clustering

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Identifying optimal no. of clusters 'k' through elbow method

wcss = []  # Within-cluster sum of squares

#On PCA data
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(df_pca)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(6, 4))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

the "elbow" point can be between 3-6, but lets plot and see how well the clustering is taking place


```python
# K-means with 3 clusters
kmeans_3 = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters_3 = kmeans_3.fit_predict(df_pca)

# Adding the cluster labels to the PCA df
df_pca['Cluster_3'] = clusters_3
```

```python
# Plotting for 3 clusters

plt.figure(figsize=(6, 4))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster_3', data=df_pca, palette='viridis', marker='o', edgecolor='k')
plt.title('K-means Clustering with 3 Clusters')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

Now , lets try with 4 clusters

```python
# K-means with 4 clusters
kmeans_4 = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters_4 = kmeans_4.fit_predict(df_pca)

# Adding the cluster labels to the PCA df
df_pca['Cluster_4'] = clusters_4
```

```python
# Plotting for 4 clusters

plt.figure(figsize=(6, 4))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster_4', data=df_pca, palette='viridis', marker='o', edgecolor='k')
plt.title('K-means Clustering with 4 Clusters')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

Lets check with 6 clusters

```python
# K-means with 6 clusters
kmeans_6 = KMeans(n_clusters=6, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters_6 = kmeans_6.fit_predict(df_pca)

# Adding the cluster labels to the PCA df
df_pca['Cluster_6'] = clusters_6

# Plotting for 6 clusters
plt.figure(figsize=(6, 4))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster_6', data=df_pca, palette='viridis', marker='o', edgecolor='k')
plt.title('K-means Clustering with 6 Clusters')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

The plot shows six distinct clusters, each represented by a different color. 
These clusters show that the K-Means algorithm has successfully partitioned the data into groups of users with similar characteristics.
Lets analyze the original features within each cluster to identify what differentiates them so we could identify potential market segments.

```python
# Cluster Analysis with 6 clusters

# Assign clusters to the original DataFrame
df['Cluster_6'] = clusters_6 

# Get the mean values of each category for each cluster
cluster_6_summary = df.groupby('Cluster_6').mean()

cluster_6_summary
```

```python
cluster_6_summary.T
```

```python
# Get the top 4 mean values for each cluster
top_4_per_cluster = cluster_6_summary.apply(lambda x: x.nlargest(4), axis=1)

for cluster in top_4_per_cluster.index:
    print(f"Cluster {cluster} Top Categories:")
    top_categories = top_4_per_cluster.loc[cluster].dropna()  # Drop NaN values
    print(top_categories)
    print("\n") 
```

The initial clustering with 6 clusters showed significant overlap, likely due to the dominance of the "chatter" category, which overshadowed other important interests.
To create clearer and more distinct segments, we will remove "chatter" and re-run the analysis using 6 clusters.

```python
df.columns.values
```

```python
# Remove the "chatter" column from the original dataset
df_no_chatter = df.drop(columns=['Cluster_6', 'chatter'])

# Re-run the PCA on the dataset without "chatter"
df_scaled_no_chatter = scaler.fit_transform(df_no_chatter.select_dtypes(include=[float, int]))
df_pca_no_chatter = pca.fit_transform(df_scaled_no_chatter)

# K-means with the new data 
kmeans_no_chatter = KMeans(n_clusters=6, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters_no_chatter = kmeans_no_chatter.fit_predict(df_pca_no_chatter)

# Assign the new clusters to the original data (without chatter)
df_no_chatter['Cluster_6'] = clusters_no_chatter

# Calculate the mean values for each category within the new clusters
cluster_6_summary_no_chatter = df_no_chatter.groupby('Cluster_6').mean()

```

```python
cluster_6_summary_no_chatter.T
```

```python
# Get the top 4 categories for each of the new clusters
top_4_per_cluster_no_chatter = cluster_6_summary_no_chatter.apply(lambda x: x.nlargest(4), axis=1)

# Display the top categories without NaNs
for cluster in top_4_per_cluster_no_chatter.index:
    print(f"Cluster {cluster} Top Categories (without chatter):")
    top_categories_no_chatter = top_4_per_cluster_no_chatter.loc[cluster].dropna()
    print(top_categories_no_chatter)
    print("\n")
```

## Correlation Analysis

```python
# Calculate the correlation matrix
correlation_matrix = df_no_chatter.corr()

# removing self correlation 
correlation_pairs = correlation_matrix.unstack()
correlation_pairs = correlation_pairs[correlation_pairs < 1]

# checking for highest correlation
correlation_pairs.sort_values(ascending=False).drop_duplicates().head(10)
```

The strong correlations between categories like personal fitness and health nutrition, as well as fashion and cooking, align with clusters focused on lifestyle and health,suggesting that these users have a well-rounded set of interests.
The correlations between parenting, religion, and sports fandom support the clusters that are centered around traditional and community-oriented values, showing that these interests often coexist.
